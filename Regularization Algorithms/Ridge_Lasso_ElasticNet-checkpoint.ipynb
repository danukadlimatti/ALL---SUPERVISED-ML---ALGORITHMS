```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
boston_pd = pd.read_csv('boston.csv')
```

```python
boston_pd
```

```python
boston_pd=boston_pd.iloc[:,1:]
boston_pd
```

```python
# input 
X = boston_pd.iloc[:, :-1] 

#output 
Y = boston_pd.iloc[:, -1] 
# boston_pd.head()
```

```python
X
```

```python
# boston_pd.to_csv('boston.csv')
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.25,random_state=1)

(x_train.shape, y_train.shape,x_test.shape, y_test.shape)
```

```python
x_train
```

```python
# Apply multiple Linear Regression Model
from sklearn.linear_model import LinearRegression
lreg = LinearRegression() # model
lreg.fit(x_train, y_train) 
```

```python
# Generate Prediction on test set 
lreg_y_pred = lreg.predict(x_test) 
```

```python
y_test
```

```python
lreg_y_pred
```

```python
lreg_y_pred.shape
```

```python
# calculating Mean Squared Error (mse) 
mean_squared_error = np.mean((lreg_y_pred - y_test)**2) 
print("Mean squared Error on test set : ", mean_squared_error)

y_pred1 = lreg.predict(x_train) 

mean_squared_error = np.mean((y_pred1 - y_train)**2) 
print("Mean Squared Error on Training set", mean_squared_error) 
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,lreg_y_pred)
```

#### # import Lasso regression from sklearn library 



```python
from sklearn.linear_model import Lasso 

# Train the model 
lasso = Lasso(alpha = 0.07) 
lasso.fit(x_train, y_train) 
y_pred1 = lasso.predict(x_test) 

# Calculate Mean Squared Error 
mean_squared_error = np.mean((y_pred1 - y_test)**2) 
print("Mean squared error on test set", mean_squared_error) 
lasso_coeff = pd.DataFrame() 
lasso_coeff["Columns"] = x_train.columns 
lasso_coeff['Coefficient Estimate'] = pd.Series(lasso.coef_) 

# print(lasso_coeff) 
```

```python
# import model 
from sklearn.linear_model import ElasticNet 


# Train the model 
e_net = ElasticNet(alpha = 0.1,l1_ratio=0.7) 
e_net.fit(x_train, y_train) 

# calculate the prediction and mean square error 
y_pred_elastic = e_net.predict(x_test) 
mean_squared_error = np.mean((y_pred_elastic - y_test)**2) 
print("Mean Squared Error on test set", mean_squared_error) 

e_net_coeff = pd.DataFrame() 
e_net_coeff["Columns"] = x_train.columns 
e_net_coeff['Coefficient Estimate'] = pd.Series(e_net.coef_) 
e_net_coeff 
```

```python
y_pred_elastic1 = e_net.predict(x_train) 
mean_squared_error = np.mean((y_pred_elastic1 - y_train)**2) 
print("Mean Squared Error on Training set", mean_squared_error) 
```
